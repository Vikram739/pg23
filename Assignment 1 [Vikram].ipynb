## Create data subsets

```python
d1 = data[['Type','Category']]
d1
```

```python
d2 = data[["Post Month","Type"]]
d2
```

## Merge Subsets by rows

```python
rowMerged = pd.concat([d1,d2])
rowMerged.shape
```

## Merge Subsets by columns

```python
colMerged = d1.merge(d2, left_index=True,right_index=True)
colMerged
```

```python
colMerged.shape
```

## Sort Dataset by Category

```python
s1 = d1.sort_values(by='Category')
s1
```

# Sort Dataset by Category in decending order

```python
s2 = d1.sort_values(by='Category', ascending=False)
s2
```
