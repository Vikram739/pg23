##  Removing null valued rows

```python
data = data.dropna()
data
```

## Creating the first column subset 

```python
column_data1=data[["Type","Category","Post Month","Post Weekday","Post Hour"]]
column_data1
```

## Creating the second column subset 

```python
column_data2=data[["Paid","Lifetime Post Total Reach","Lifetime Post Consumers"]]
column_data2
```

## Creating the first row subset

```python
row_data1=data.iloc[1:11]
row_data1
```

## Creating the second row subset 

```python
row_data2=data.iloc[[1,2,3,7,11,13,14,15,16,17]]
row_data2
```

## Merging two subsets

```python
m1 = row_data1.merge(row_data2)
m1
```

## Concatenating two subsets 

```python
c1 = pd.concat([row_data1,row_data2])
c1
```

## Sorting the subset 

```python
sort1= c1.sort_values(by="Post Weekday")
sort1
```

## Transposing a subset 

```python
transpose=sort1.transpose()
transpose
```

```python
print(sort1.shape)
print(transpose.shape)
```

## Reshaping the subset

```python
reshape=sort1.melt(id_vars="Type",value_vars=["Lifetime Post Total Reach","Post Weekday"])
reshape
```
