## Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Reading the data 

```python
data = pd.read_csv(r"D:\DYP\6th Sem\LAB\DSBDA\pe\airquality.csv")
```

```python
data
```

## Replacing the null values by mean values 

```python
data["Ozone"] = data["Ozone"].fillna(data["Ozone"].mean())
data["Solar.R"] = data["Solar.R"].fillna(data["Solar.R"].mean())
```

```python
data
```

## Pie Chart 

```python
plt.pie([data["Ozone"].mean(),data["Solar.R"].mean(),data["Wind"].mean(),data["Temp"].mean()], autopct="%1.2f%%", explode=[0.05,0.01,0.01,0.01], labels=["Ozone", "Solar.R", "Wind", "Temp"])
plt.plot()
```

## Bar Plot

```python
h=data.iloc[1:21,4]
h
```

```python
y_pos=np.arange(len(h))
v=range(1,21)
print(y_pos)
print(v)
```

```python
plt.bar(y_pos,height=h,width=0.9,bottom=None,data=None)
plt.xticks(y_pos,v)
plt.savefig('plot2.png')
plt.title('Temperature of 1st-20th May')
plt.xlabel('Days')
plt.ylabel('Range')
plt.show()
```

## Box Plot

```python
plt.boxplot(x=data[["Temp", "Solar.R", "Wind", "Ozone"]], labels=["Temp", "Solar.R", "Wind", "Ozone"])
plt.plot()
```

## Histogram 

```python
plt.hist(x=data["Temp"], bins=20, color="orange", label="Temp")
plt.title("Temperature count")
plt.xlabel("Temperature")
plt.ylabel("Count")
plt.plot()
```

## Line Graph

```python
plt.plot(data.iloc[0:30, 6], data.iloc[0:30, 3], color="purple")
plt.title("Wind of May month")
plt.xlabel("Day")
plt.ylabel("Wind")
```

## Scatter Plot 

```python
plt.scatter(data.iloc[0:30, 6], data.iloc[0:30, 3], color="red")
plt.title("Wind measure of May month")
plt.xlabel("Day")
plt.ylabel("Wind")
```

## Pair Plot 

```python
import seaborn

seaborn.pairplot(data.iloc[:,[1,2,3,4,7]],hue ='humidity' )
plt.show()
```

## Heatmap

```python

x = data.pivot_table(index='Day',columns='Month',values='Temp',aggfunc='sum')
x
```

```python
data['Month'].unique()
```

```python
seaborn.heatmap(x,xticklabels=['Feb','Mar','Apr','May','Jun'],annot=True)
plt.show()
```
